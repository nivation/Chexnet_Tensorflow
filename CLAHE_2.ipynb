```python
bgr = cv2.imread('/home/b05501009/Jubo/7_30/CropedImages/2711.jpg')
print()
img = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
lab = cv2.cvtColor(bgr, cv2.COLOR_BGR2LAB)
print(lab.shape)
l,a,b = cv2.split(lab)
clahe = cv2.createCLAHE(clipLimit=1.0,tileGridSize=(10,10))
l = clahe.apply(l)
a = clahe.apply(a)
b = clahe.apply(b)
l = np.expand_dims(l, axis=2)
a = np.expand_dims(a, axis=2)
b = np.expand_dims(b, axis=2)

lab = np.concatenate((np.array(l),np.array(a),np.array(b)),axis = 2)
print(lab.shape)
bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
img = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
bgr = cv2.imread('/home/b05501009/Jubo/7_30/CropedImages/1332.jpg')
print(bgr)
print(bgr.shape)
print(type(bgr))
img = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

lab = cv2.cvtColor(bgr, cv2.COLOR_BGR2LAB)
lab_planes = cv2.split(lab)
print(np.array(lab_planes).shape)
print(lab_planes[0])
clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(10,10))
lab_planes[0] = clahe.apply(lab_planes[0])

lab = cv2.merge(lab_planes)
bgr = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
img = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
